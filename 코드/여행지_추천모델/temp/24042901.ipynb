```python
import os
data_list = os.listdir("raw_csvs//")
data_list
```

```python
def get_province(row):
    try:
        return row.split(' ')[0]
    except:
        return ''
def get_city(row):
    try:
        return row.split(' ')[1]
    except:
        return ''

df_total = pd.DataFrame()

for data in data_list:
    print(data)
    df = pd.read_csv("raw_csvs/"+data)
    
    df["힐링_int"] = df["힐링"].replace({'O':1, 'X':0})
    df["혼자_int"] = df["혼자"].replace({'O':1, 'X':0})
    df["자연_int"] = df["자연"].replace({'O':1, 'X':0})
    df["액티비티_int"] = df["액티비티"].replace({'O':1, 'X':0})
    
    df_data = df[["name", "addr1", "cat1", "cat2", "cat3", "힐링_int", "혼자_int", "자연_int", "액티비티_int"]]
    df_data["province"] = df_data["addr1"].apply(get_province)
    df_data["city"] = df_data["addr1"].apply(get_city)
    df_data.to_csv("review_data/"+data.replace(".csv", "_data.csv"))
    
    df_total = pd.concat([df_total, df_data])
```

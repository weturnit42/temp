```python
import pandas as pd
import os
```

```python
lower_traveler_region = open("lower_traveler_region.txt", "r", encoding="utf-8")
temp = lower_traveler_region.readlines()
# print(lower_traveler_region.readlines())
```

```python
lower_traveler_region_list = [low_region.replace('\n', '') for low_region in temp]
lower_traveler_region_list
```

```python
house_total_score = pd.read_csv("total_score/house_total_score.csv")
house_total_score["dummy"] = 1
```

```python
cond_house_total_score_low_region = house_total_score["city"].isin(lower_traveler_region_list)
house_total_score_low_region = house_total_score.loc[cond_house_total_score_low_region]
```

```python
house_pivot = house_total_score_low_region.pivot_table(index="cat3", values="dummy", aggfunc="sum").reset_index()
```

```python
house_total_score = pd.read_csv("total_score/restaurant_total_score.csv")
house_total_score["dummy"] = 1

cond_house_total_score_low_region = house_total_score["city"].isin(lower_traveler_region_list)
house_total_score_low_region = house_total_score.loc[cond_house_total_score_low_region]

restaurant_pivot = house_total_score_low_region.pivot_table(index="cat3", values="dummy", aggfunc="sum").reset_index()
restaurant_pivot
```

```python
house_total_score = pd.read_csv("total_score/tour_place_total_score.csv")
house_total_score["dummy"] = 1

cond_house_total_score_low_region = house_total_score["city"].isin(lower_traveler_region_list)
house_total_score_low_region = house_total_score.loc[cond_house_total_score_low_region]

tour_place_pivot = house_total_score_low_region.pivot_table(index="cat3", values="dummy", aggfunc="sum").reset_index()
tour_place_pivot
```

```python
total_pivot = pd.DataFrame()

total_pivot = pd.concat([total_pivot, house_pivot])
total_pivot = pd.concat([total_pivot, restaurant_pivot])
total_pivot = pd.concat([total_pivot, tour_place_pivot])

total_pivot.sort_values(by="dummy", ascending=False)
```

```python
df_codes = pd.read_csv("search_score/df_codes.csv")
df_codes
```

```python
total_pivot_decoded = pd.merge(total_pivot, df_codes, left_on="cat3", right_on="code", how="left")
total_pivot_decoded
```

```python
total_pivot_decoded.sort_values(by="dummy", ascending=False)[["cat3", "dummy", "subcategory"]].to_csv("total_pivot_decoded.csv")
```

```python
total_pivot_decoded_v2 = total_pivot_decoded.sort_values(by="dummy", ascending=False)[["cat3", "dummy", "subcategory"]]
total_pivot_decoded_v2
```

```python
def to_cat2(row):
    return row[:-4]

def to_cat1(row):
    return row[:-6]

total_pivot_decoded_v2["cat2"] = total_pivot_decoded_v2["cat3"].apply(to_cat2)
total_pivot_decoded_v2["cat1"] = total_pivot_decoded_v2["cat3"].apply(to_cat1)
total_pivot_decoded_v2
```

```python
codes = open("search_score/search_dict.txt", 'r', encoding="utf-8")
codes_temp = (codes.readlines())

df_codes_v2 = pd.DataFrame()

for i, temp in enumerate(codes_temp):
#     print(temp.split('\t'))
    df_codes_v2[i] = temp.split('\t')
    
df_codes_v2 = df_codes_v2.T
df_codes_v2
```

```python
df_codes_v2["cat2"] = df_codes_v2[0].apply(to_cat2)
df_codes_v2["cat1"] = df_codes_v2[0].apply(to_cat1)
df_codes_v2
```

```python
total_pivot_decoded_v3 = pd.merge(total_pivot_decoded_v2, df_codes_v2[[1, 2, 0]], left_on="cat3", right_on=0, how="left")
total_pivot_decoded_v4 = total_pivot_decoded_v3[["subcategory", 2, 1, "dummy"]]
total_pivot_decoded_v4 
```

```python
category_pivot = total_pivot_decoded_v4.pivot_table(index=2, values="dummy", aggfunc="sum")
category_pivot = category_pivot.reset_index().sort_values(by="dummy", ascending=False)
category_pivot.to_csv("category_pivot.csv")
```

```python
main_category_pivot = total_pivot_decoded_v4.pivot_table(index=1, values="dummy", aggfunc="sum")
main_category_pivot = main_category_pivot.reset_index().sort_values(by="dummy", ascending=False)
main_category_pivot.to_csv("main_category_pivot.csv")
```

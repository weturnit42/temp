```python
import torch
from transformers import BertTokenizer, BertModel
import torch.nn.functional as F

# BERT 토크나이저와 모델 로드
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained('bert-base-multilingual-cased')

# 유사도 계산 및 결과 기록을 위한 함수 정의
def calculate_similarity(sentence, keywords):
    try:
        # 문장 토큰화 및 BERT 인코딩
        sentence_inputs = tokenizer(sentence, return_tensors="pt")
        sentence_outputs = model(**sentence_inputs)
        sentence_vector = sentence_outputs.last_hidden_state.mean(dim=1)  # 문장 벡터 계산

        similarities = {}

        for keyword in keywords:
            # 키워드 토큰화 및 BERT 인코딩
            keyword_inputs = tokenizer(keyword, padding=True, truncation=True, return_tensors="pt")
            keyword_outputs = model(**keyword_inputs)
            keyword_vector = keyword_outputs.last_hidden_state.mean(dim=1)  # 키워드 벡터 계산

            # 유사도 계산
            similarity = F.cosine_similarity(sentence_vector, keyword_vector)
            similarities[keyword] = similarity.item()

        return similarities
    except Exception as e:
        print(f"Error occurred: {e}")
        return {keyword: 0 for keyword in keywords}

# 데이터프레임의 모든 리뷰에 대해 유사도 계산하고 결과를 데이터프레임에 추가하는 함수 정의
def update_dataframe(df, keywords):
    for idx, row in df.iterrows():
        sentence = row["review"]
        similarities = calculate_similarity(sentence, keywords)
        for keyword, similarity in similarities.items():
            df.at[idx, keyword] = similarity

# 새로운 컬럼 추가를 위한 키워드 정의
keywords = ['힐링하기 좋아요', '독특하고 특색있어요', '자연 경관이 좋아요', '액티비티가 다양해요']

# 데이터프레임의 모든 리뷰에 대해 유사도 계산하고 결과를 업데이트
update_dataframe(df6, keywords)
```
